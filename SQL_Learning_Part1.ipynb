## Creating helper function for extracting data from databses using SQL query

```
def extract_data(SQL_query):
  """
  Docstring:

  Input: SQL query
  Output: Pandas Dataframe of data extracted
  """
  myDB = MySQLdb.connect(db='classicmodels')
  query_result=pd.read_sql(SQL_query,con=myDB)
  myDB.close()
  return query_result
```

# Section 4: Filtering data

## Using Distinct

```
extract_data('select distinct jobTitle from employees;')
```

## Using or

```
extract_data('select CONCAT(firstName, " ",lastName) as full_name	,jobTitle,officeCode \
from employees \
where jobTitle="Sales Rep" or  jobTitle = "Sales Manager (NA)" \
order by employeeNumber ASC;')
```

## Using 'In'

```
extract_data('select CONCAT(firstName, " ",lastName) as full_name	,jobTitle,officeCode \
from employees \
where jobTitle in ("Sales Rep", "Sales Manager (NA)", "VP Sales") ;')
```

# Section 5. Joining tables

## Using Concat

```
extract_data('select CONCAT(contactFirstName, " ", contactLastName) as contact, country from customers;')
```

## Inner Join

```
extract_data('select productCode, productName, textDescription \
from products t1 INNER JOIN productlines t2 \
ON t1.productline = t2.productline;')
```

## Left Join

```
extract_data('select firstName, lastName,checkNumber, amount \
from employees \
LEFT JOIN customers on employeeNumber = salesRepEmployeeNumber \
LEFT JOIN payments on payments.customerNumber = customers.customerNumber ORDER BY checkNumber;')
```

## Right Join

```
extract_data("select employeeNumber,customerNumber from customers right join employees on salesRepEmployeeNumber = employeeNumber where customerNumber;")
```

# Section 10: Modifying data in MySQL

## Creating table

```
myDB = MySQLdb.connect(db='classicmodels')
cursor = myDB.cursor()
query='create table if not exists \
tasks(taskid int auto_increment,title varchar(250) not null,\
start_date date,due_date date, \
priority tinyint not null default 3, \
description text,primary key(taskid))'
cursor.execute(query)
myDB.commit()
myDB.close()
```

```
extract_data('select * from tasks')
```

## Inserting data into table

```
myDB = MySQLdb.connect(db='classicmodels')
cursor = myDB.cursor()
query="insert into tasks(title,priority) values('My first task',1),('it is the second task',2),('this is the third task',3)"
cursor.execute(query)
myDB.commit()
myDB.close()
```

```
extract_data("select * from tasks")
```

```
extract_data("show tables") # task can be seen in tables
```

```
myDB = MySQLdb.connect(db='classicmodels')
cursor = myDB.cursor()
query='create table if not exists stats(totalProduct int,totalCustomer int,totalOrder int)'
cursor.execute(query)
myDB.commit()
myDB.close()
```

```
extract_data("select * from stats")
```

```
myDB = MySQLdb.connect(db='classicmodels')
cursor = myDB.cursor()
query="insert into stats(totalProduct,totalCustomer,totalOrder) values((select count(*) from products),(select count(*) from customers),(select count(*) from orders))"
cursor.execute(query)
myDB.commit()
myDB.close()
```
